```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
df['subclass'] = encoder.fit_transform(df[['subclass']]).astype(int)
df['class'] = encoder.fit_transform(df[['class']]).astype(int)

# Splitting Features and Target
X = df.drop(columns=['subclass'])  # Features
y = df['subclass']  # Target variable
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.feature_selection import VarianceThreshold
from imblearn.over_sampling import SMOTE
```

```python
# Handling outliers using the IQR method for the 'u' column
Q1 = df['u'].quantile(0.25)
Q3 = df['u'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Removing outliers
filtered_df = df[(df['u'] >= lower_bound) & (df['u'] <= upper_bound)]

# Ensure 'y' is filtered accordingly
filtered_y = y.loc[filtered_df.index]  # Ensure y matches filtered_df

# Remove constant features
var_thresh = VarianceThreshold(threshold=0.0)  # Threshold 0 removes constant features
X_filtered = var_thresh.fit_transform(filtered_df.drop(columns=['subclass']))
filtered_feature_names = filtered_df.drop(columns=['subclass']).columns[var_thresh.get_support()]

# Feature Selection using SelectKBest
selector = SelectKBest(score_func=f_classif, k=10)
X_new = selector.fit_transform(X_filtered, filtered_y)  # Use filtered_y
selected_features = filtered_feature_names[selector.get_support()]

print("Selected Features:", selected_features)

# Balancing dataset using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_new, filtered_y)  # Use filtered_y

print("Class distribution after SMOTE:")
print(pd.Series(y_resampled).value_counts())

# Boxplot after handling outliers
plt.figure(figsize=(6, 4))
sns.boxplot(x=filtered_df['u'])
plt.title('Boxplot after Handling Outliers in u Column')
plt.show()
```

```python

from sklearn.preprocessing import StandardScaler
# Standardizing the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)
```

```python
from sklearn.model_selection import train_test_split
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_resampled, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
#Decision Tree Classifier Model
dt_m = DecisionTreeClassifier()
dt_m.fit(X_train, y_train)
y_pred= dt_m.predict(X_test)
acc_dt=accuracy_score(y_test,y_pred)
c_dt= classification_report(y_test,y_pred)
print("Accuracy for dt: ",acc_dt)
print(c_dt)
```

```python
#Logistic Regression Model
lr_m = LogisticRegression()
lr_m.fit(X_train, y_train)
y_pred= lr_m.predict(X_test)
acc_lr=accuracy_score(y_test,y_pred)
c_lr= classification_report(y_test,y_pred)
print("Accuracy for lr: ",acc_lr)
print(c_lr)
```

```python
#Random Forest Classifier Model
rf_m = RandomForestClassifier()
rf_m.fit(X_train, y_train)
y_pred= rf_m.predict(X_test)
acc_rf=accuracy_score(y_test,y_pred)
c_rf= classification_report(y_test,y_pred)
print("Accuracy for rf: ",acc_rf)
print(c_rf)
```
